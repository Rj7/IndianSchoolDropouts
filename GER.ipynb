```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt

import statsmodels.api as sm
```

```python
from statsmodels.graphics.api import qqplot
```

```python
ger = pd.read_csv("ger_primary_upper.csv")
```

```python
table = pd.DataFrame(ger, columns=["Level","Primary-Total", "Upper-Primary-Total" ])
```

```python
level = table["Level"]
```

```python
del table["Level"]
```

```python
dum = []
for item in level:
    dum.append(item)
```

```python
table.index =pd.Index(dum)
```

```python
table.plot(figsize=(12,8))
```

```python
table
```

```python
plt.show()
```

```python
secondary = pd.read_csv("ger_secondary_high.csv")
```

```python
secondary_df = pd.DataFrame(secondary, columns=["Level","Secondary-Total", "Senior-Secondary-Total", "Higher-Education-Total"])
```

```python
level = secondary_df["Level"]
```

```python
sec_dum = []
for item in level:
    sec_dum.append(item)
```

```python
del secondary_df["Level"]
```

```python
secondary_df.index = pd.Index(sec_dum)
```

```python
school_education = pd.concat([table, secondary_df], axis=1 )
```

```python
school_education.plot(figsize=(12,8))
```

```python
school_education_2005 = school_education.tail(9)
```

```python
school_education_2005
```

```python
school_education_2005.plot(figsize=(12,8))
```
